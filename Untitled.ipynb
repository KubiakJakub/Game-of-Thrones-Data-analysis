```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
GoT = pd.read_csv('got_data.csv')
GoT.head()
```

```python
GoT.info()
```

```python
plt.figure(figsize=(30,8))
sns.barplot(x='Season', y='Imdb Rating', data=GoT)
sns.set(font_scale=2)
plt.tight_layout()
```

```python
plt.figure(figsize=(40,30))
sns.heatmap(GoT.corr(), annot=True)
sns.set(font_scale=3)
```

```python
plt.figure(figsize=(30,8))
sns.set(font_scale=3)
sns.barplot(x='Season', y='Notable Death Count', data=GoT)
plt.tight_layout()
```

```python
sns.set(font_scale=1.5, style='whitegrid')
sns.pairplot(GoT)
```

```python
seasonRating = pd.DataFrame(GoT[['Season', 'Imdb Rating']])
avg_s_rtg = seasonRating.groupby(by="Season").mean()
avg_s_rtg
```
